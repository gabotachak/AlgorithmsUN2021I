 ####  **2-1: Fibonacci Number**  #### 

```python
#include <iostream>
#include <cassert>

using namespace std;
int fibonacci_naive(int n) {
    if (n <= 1)
        return n;

    return fibonacci_naive(n - 1) + fibonacci_naive(n - 2);
}

int fibonacci_fast(int n) {
    int str[n+1];
    str[0]=0;
    str[1]=1;
    for(int i=2; i<=n; i++ ){
    	str[i]= str[i-1]+ str[i-2];
	}
    return str[n];
}

void test_solution() {
    assert(fibonacci_fast(3) == 2);
    assert(fibonacci_fast(10) == 55);
    for (int n = 0; n < 20; ++n)
        assert(fibonacci_fast(n) == fibonacci_naive(n));
}

int main() {
    int n = 0;
    cin >> n;
    test_solution();
    cout << fibonacci_fast(n) << '\n';
    return 0;
}
```

 #### **2-3: Greatest Common Divisor** #### 

```python
#include <iostream>

using namespace std;
int gcd_naive(int a, int b) {
  int current_gcd = 1;
  for (int d = 2; d <= a && d <= b; d++) {
    if (a % d == 0 && b % d == 0) {
      if (d > current_gcd) {
        current_gcd = d;
      }
    }
  }
  return current_gcd;
}

int gcd(int a, int b){
	int current_gcd = 1;
	int mayor,menor;
	if(a>=b){
		mayor = a;
		menor = b;	
	}else{
		mayor = b;
		menor = a;	
	}

    for(int i= 0; current_gcd!=0;i++){
		int c = mayor/menor;
		current_gcd = mayor - menor*c;
		mayor = menor;
		menor = current_gcd;
    }
	
	return mayor;
}

int main() {
  int a, b;
  cin >> a >> b;
  cout <<gcd(a, b) << endl;
  return 0;
}
```

 #### **2-7: Last Digit of the Sum of Fibonacci Numbers Again** #### 

```python
#include <iostream>
#include <vector>
using std::vector;

long long get_fibonacci_partial_sum_naive(long long from, long long to) {
    long long sum = 0;

    long long current = 0;
    long long next  = 1;

    for (long long i = 0; i <= to; ++i) {
        if (i >= from) {
            sum += current;
        }

        long long new_current = next;
        next = next + current;
        current = new_current;
    }

    return sum % 10;
}


int last_digit(long long n){
	if(n<=1){
		return n;
	}
	
	n = n%60;
	
	long long previous = 0;
    long long current  = 1;

	
    for(int i=2; i<=n+2; i++ ){
    	long long tmp_previous = previous%10;
        previous = current%10;
        current = (tmp_previous + current)%10;
	}
	
	if(current == 0){
		return 9;
	}else{
		return current-1;
	}           
    
}


int last_digit_2(long long m, long long n){
	if(n<=1){
		return n;
	}
	
	int ln =  last_digit(n);
	int lm = 0;
	if(m < 2){
		lm = 0;
	}else{
		lm = last_digit(m-1); 
	}
	
	/*std::cout<<"ln = "<<ln<<", lm= "<<lm<<"\n";*/
	
	if(ln<lm){
		return 10+ln-lm;
	}else{
		return ln-lm;
	}
	  
}


int main() {
    long long from, to;
    std::cin >> from >> to;
    std::cout << last_digit_2(from, to) << '\n';
}
```
